```python
import numpy as np
import subprocess
from scipy import signal
import skimage.measure
```

```python
def softmax(a):
    a=np.exp(a)
    a_sum=np.sum(a)
    b=np.array([np.float32(x/a_sum) for x in a])
    return b
def sigmoid(a):
    b=np.array([np.float32(np.exp(x)/(np.exp(x)+1)) for x in a])
    return b
```

epsilon=0.01
num_iteration=40

```python
def check(c,c1):
    for num1,num2 in zip(c,c1):
        if(np.abs(num1-num2)>epsilon):
            print "Not working"
            print str(num1)+" "+str(num2)+" "+str(np.abs(num1-num2))
```

```python
#testing convolution with multiplication
subprocess.call(["make"])
for i in range(num_iteration):
    size1=np.random.randint(10,100)
    size2=size1//np.random.randint(2,10)
    padding=np.random.randint(0,20)
    a=np.random.randn(size1,size1).astype(np.float32)
    b=np.random.randn(size2,size2).astype(np.float32)
    a_columMajor=a.T
    b_columnMajor=b.T
    f = open("input.txt", "w")
    for elem in a_columMajor:
        f.write(" ".join(str(num) for num in elem))
        f.write(" ")
    f.close();
    f=open("kernel.txt","w")
    for elem in b_columnMajor:
        f.write(" ".join(str(num) for num in elem))
        f.write(" ")
    f.close()
    option=np.random.randint(1,3)
    if(option==1):
        subprocess.call(["./main","convolution",str(padding),"input.txt",str(size1),"kernel.txt",
                     str(size2),"output.txt"])
    else:
        subprocess.call(["./main","convolution_mult",str(padding),"input.txt",str(size1),"kernel.txt",
                     str(size2),"output.txt"])
    f=open("output.txt","r")
    line=f.readline()
    c=np.array([np.float32(x) for x in line.split()])
    f.close()
    a1=np.pad(a, (padding,padding), 'constant', constant_values=(0, 0))
    c1=np.squeeze(signal.correlate2d(a1,b,mode="valid").T.reshape(-1,1))
    
    check(c,c1)
```

```python
#testing activation
subprocess.call(["make"])
for i in range(num_iteration):
    size1=np.random.randint(1,100)
    size2=np.random.randint(1,100)
    a=np.random.randn(size1,size2).astype(np.float32)
    f = open("input.txt", "w")
    for elem in a.T:
        f.write(" ".join(str(num) for num in elem))
        f.write(" ")
    f.close();
    option=np.random.randint(1,3)
    if(option==1):
        c1=np.squeeze(np.maximum(a,0).T.reshape(-1,1))
        subprocess.call(["./main","relu","input.txt",str(size1),str(size2),"output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
    else:
        c1=np.squeeze(np.tanh(a).T.reshape(-1,1))
        subprocess.call(["./main","tanh","input.txt",str(size1),str(size2),"output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
```

```python
#testing pooling
subprocess.call(["make"])
# num_iteration=1
for i in range(num_iteration):
    size1=np.random.randint(10,30)
    factors=[]
    for i in range(1,size1/2):
        if(size1%i==0):
            factors.append(i);
    size2_index=np.random.randint(0,len(factors))
    size2=factors[size2_index]
#     size1=15
#     size2=5
    a=np.random.randn(size1,size1).astype(np.float32)
    a_columMajor=a.T
    f = open("input.txt", "w")
    for elem in a_columMajor:
        f.write(" ".join(str(num) for num in elem))
        f.write(" ")
    f.close();
   
    option=np.random.randint(1,3)
    if(option==1):
#         print("max_pool "+"input size "+str(size1)+" kernel size "+str(size2))
        c1=np.squeeze(skimage.measure.block_reduce(a, (size2,size2), np.max).T.reshape(-1,1))
        subprocess.call(["./main","maxpool","input.txt",str(size1),str(size2),"output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
    else:
#         print("avg_pool "+"input size "+str(size1)+" kernel size "+str(size2))
        c1=np.squeeze(skimage.measure.block_reduce(a, (size2,size2), np.mean).T.reshape(-1,1))
        subprocess.call(["./main","avgpool","input.txt",str(size1),str(size2),"output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
```

```python
#testing sigmoid and softmax
subprocess.call(["make"])
for i in range(num_iteration):
    size1=np.random.randint(1,100)
    a=np.random.randn(size1).astype(np.float32)
    f = open("input.txt", "w")
    f.write(" ".join(str(num) for num in a))
    f.close();
    
    option=np.random.randint(1,3)
    if(option==1):
#         print("softmax "+"input size "+str(size1))
        c1=np.squeeze(softmax(a))
        subprocess.call(["./main","softmax","input.txt","output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
    else:
#         print("sigmoid "+"input size "+str(size1))
        c1=np.squeeze(sigmoid(a))
        subprocess.call(["./main","sigmoid","input.txt","output.txt"])
        f=open("output.txt","r")
        line=f.readline()
        c=np.array([np.float32(x) for x in line.split()])
        f.close()
        check(c,c1)
```
